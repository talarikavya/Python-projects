#  What is the average price of cars across different brands, and which brands have the highest and lowest average prices

```python
ft.head()
```

```python
df = df.dropna(subset=['Price', 'Brand'])   #here rows are used in dataset
```

```python
avg_brand = df.groupby('Brand')['Price'].mean().reset_index()
avg_price_brand = avg_brand.sort_values(by='Price', ascending=False)
```

```python
highest_avg_price = avg_price_brand.iloc[0]
lowest_avg_price = avg_price_brand.iloc[-1] 
```

```python
print("highest average price:")
print("lowest average price:")
print(highest_avg_price)
print(lowest_avg_price)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_price_brand, x='Brand', y='Price', palette='viridis')
plt.title("Average Car Price Across Different Brands")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=60)  # Rotate brand names for better readability
plt.show()
```
