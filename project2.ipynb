```python
# what are the most common ratings and how are they distributed across different types of content

rating_counts = dt['rating'].value_counts()
print("common ratings:\n", rating_counts)
```

```python
content_rating = dt.groupby('type')['rating'].value_counts().fillna(0)
print("different types of content:\n", content_rating)
```

```python
content_rating.plot(kind='bar')
plt.title('distributed across different content rating')
plt.xlabel('content_rating')
plt.ylabel('rating_counts')
plt.legend()
plt.show()
```

```python
#find which types of genre has most movies and tv shows

movies_df = dt[dt['type'] == 'Movie']
tv_shows_df = dt[dt['type'] == 'TV Show']
```

```python
movies_genres = movies_df['listed_in'].str.split(',').explode().str.strip()
tv_shows_genres = tv_shows_df['listed_in'].str.split(',').explode().str.strip()
```

```python
print("genres in TV Shows:")
print(movies_genres.head(10))
print(tv_shows_genres.head(10))
```
