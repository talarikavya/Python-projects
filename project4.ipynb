# Which countries or regions have the highest and lowest total confirmed cases and fatalities

```python
set.head()
```

```python
region_data = set.groupby('Country/Region')[['ConfirmedCases', 'Fatalities']].sum()
```

```python
highest_confirmed = region_data['ConfirmedCases'].idxmax()     #here idmax is defining the highest cases
highest_fatalities = region_data['Fatalities'].idxmax()
```

```python
lowest_confirmed = region_data['ConfirmedCases'].idxmin()    #here idmin is defining the lowest cases
lowest_fatalities = region_data['Fatalities'].idxmin()
```

```python
print(f"region with highest confirmed cases: {highest_confirmed}")
print(f"region with lowest confirmed cases: {lowest_confirmed}")
print(f"region with highest fatalities cases: {highest_fatalities}")
print(f"region with lowest confirmed cases: {lowest_confirmed}")
```

# Is there a correlation between temperature (temp) and the number of confirmed cases or fatalities

```python
set.head()
```

```python
df_clean = set.dropna(subset=['temp', 'ConfirmedCases', 'Fatalities'])
```

```python
correlation_temp_cases = df_clean['temp'].corr(df_clean['ConfirmedCases'])
correlation_temp_fatalities = df_clean['temp'].corr(df_clean['Fatalities'])
```

```python
print(f"Correlation between temperature and confirmed cases: {correlation_temp_cases}")
print(f"Correlation between temperature and fatalities: {correlation_temp_fatalities}")
```
