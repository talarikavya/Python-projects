# Which users have the highest average engagement (likes and retweets per post)

```python
data['Likes'] = data['Likes'].fillna(0)        #here assuming of columns of likes and retweets
data['Retweets'] = data['Retweets'].fillna(0)
```

```python
data['total_engagement'] = data['Likes']+data['Retweets'] #here storing of total engagement
```

```python
user_engagement = data.groupby('User')[['Likes','Retweets','total_engagement']].mean()
```

```python
user_engagement_sorted = user_engagement.sort_values('total_engagement', ascending=True)
```

```python
print(user_engagement_sorted.head())
```
