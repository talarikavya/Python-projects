# Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit

```python
pro.head()
```

```python
pro[['Size (cm)', 'Weight (g)']].isnull().sum()
```

```python
pro_clean = pro.dropna(subset=['Size (cm)', 'Weight (g)'])
```

```python
correlation = pro_clean[['Size (cm)', 'Weight (g)']].corr()  #corr is size and weight of colunms
```

```python
print(correlation)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Size (cm)', y='Weight (g)', data=pro_clean)
plt.title("Relationship Between Size and Weight of Fruit")
plt.xlabel("Size (cm)")
plt.ylabel("Weight (g)")
plt.show()
```

#  How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity
