# What are the trends in total purchase amount over time (based on Purchase Date).Are there any seasonal patterns

```python
ds.head()
```

```python
ds.info()
```

```python
ds['Purchase Date'] = pd.to_datetime(ds['Purchase Date'], errors='coerce')
```

```python
ds = ds.dropna(subset=['Purchase Date'])
```

```python
ds['YearMonth'] = ds['Purchase Date'].dt.to_period('M')
```

```python
monthly_purchase = ds.groupby('YearMonth')['Total Purchase Amount'].sum().reset_index()
monthly_purchase.head()
```

```python
ds['Month'] = ds['Purchase Date'].dt.month
ds['Year'] = ds['Purchase Date'].dt.year
```

```python
monthly_avg_purchase = ds.groupby('YearMonth')['Total Purchase Amount'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_avg_purchase, x='YearMonth', y='Total Purchase Amount')  
plt.title('Average Purchase Amount by Month (Seasonal Pattern)')
plt.xlabel('Month')
plt.ylabel('Average Purchase Amount')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

# Which product categories generate the highest revenue and which are the most popular based on quantity purchased

```python
ds.head()
```

```python
ds['Revenue'] = ds['Total Purchase Amount'] * ds['Quantity']
```

```python
revenue_categories = ds.groupby('Product Category')['Quantity'].mean().reset_index()
```

```python
revenue_categories = revenue_categories.sort_values(by='Quantity', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=revenue_categories, x='Product Category', y='Quantity', palette='viridis')
plt.title('Revenue Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
