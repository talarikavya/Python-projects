```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\Talar\OneDrive\Desktop\python\python\StudentPerformanceFactors.csv')
```

```python
#is there a siginificant relationship between class attendance and exam scores 
attendance = df['Attendance']   #assuming the attendace and examscore in given dataset
exam_scores = df['Exam_Score']
```

```python
plt.scatter(attendance, exam_scores)  #relationship of class attendance and exam score in visualize
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.title('Class Attendance vs Exam Scores')
plt.show()
```
